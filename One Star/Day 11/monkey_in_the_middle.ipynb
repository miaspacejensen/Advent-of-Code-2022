```python
import math
```

```python
def get_data(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    lines = text.split("\n\n")
    return lines
```

```python
def operation(op, old_item):
    op_split = op.split("old")
    new_op = str(old_item).join(op_split)
    new = eval(new_op)
    return new

def test_divisible(item, test):
    if item % test == 0: 
        return True
    else: 
        return False
    
def create_monkey_dic(monkeys):
    m_dic = {}
    for i, monkey in enumerate(monkeys):

        id, start_items, op, test, true_outcome, false_outcome = monkey.rsplit("\n")

        # starting items
        start_items = start_items.rsplit(": ")[1].rsplit(", ")
        start_items = [int(item) for item in start_items]

        # operation
        op = op.rsplit(": new = ")[1]

        # test
        test = int(test.split("divisible by ")[1])

        # outcomes
        true_outcome = int(true_outcome.split(": throw to monkey ")[1])
        false_outcome = int(false_outcome.split(": throw to monkey ")[1])

        m_dic[i] = [start_items, op, test, true_outcome, false_outcome]

    return m_dic
```

```python
monkeys = get_data("input.txt")
m_dic = create_monkey_dic(monkeys)
count_dic = {}
m_dic_start = create_monkey_dic(monkeys)
current_levels = {}
i = 0
rounds = 20
round = 1
while True:

    # print()
    # for key, val in m_dic.items():
    #     print(f"Monkey {key}: {val[0]}")
    # print()

    key, val = i, m_dic.get(i)
    print("Monkey", key)

    worry_levels, op, test, if_true, if_false = val

    try:
        count_dic[i] += len(worry_levels)
    except:
        count_dic[i] = len(worry_levels)

    inspected = worry_levels.copy()
    for worry in worry_levels:
        new_worry = math.floor(operation(op, worry) / 3)
        
        test_result = test_divisible(new_worry, test)
        if test_result == True:
            dest_monkey = if_true
            
        else:
            dest_monkey = if_false

        print(" ", new_worry, "\tto Monkey", dest_monkey)

        dest_worry_levels = m_dic[dest_monkey][0]
        dest_worry_levels.append(new_worry)

    m_dic[i][0] = []

    i+=1
    if i == len(m_dic):
        print(f"\n---------End Round {round}---------\n")
        for key, val in m_dic.items():
            print(f"Monkey {key}: {val[0]}")
        print(f"\n-----------------------------\n")
        
        round += 1
        i = 0
        if round == rounds + 1:
            break
```

```python
print(f"Total number of times each monkey inspects items:\n{count_dic}\n")

values = sorted(list(count_dic.values()))
result = values[-2] * values[-1]
print("Result =", values, result)
```
